```python
#extract csv file from zipfile
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall() #extract the zip file
zip_ref.close()#close the file
```

```python
#read data from the file and handle null values
import pandas as pd
df = pd.read_csv('orders.csv', na_values = ['Not Available', 'unknown'])
#df.head(20)
df['Ship Mode'].unique()
```

```python
#Unique Cities
df['City'].unique()
```

```python
#df.columns # - displays the columns
df.columns = df.columns.str.lower() #converts the columns to lower case
```

```python
df.columns # - displays the columns
```

```python
df.columns = df.columns.str.replace(' ','_') # - replaces space with underscore
```

```python
df.columns # - displays the columns
df
```

```python
#derive new columns like discount, sale price and profit
df['discount'] = df['list_price'] * df['discount_percent']*0.01
df['sale_price'] = df['list_price'] - df['discount']
df['profit'] = df['sale_price'] - df['cost_price']
```

```python
df
```

```python
#df.dtypes
#convert the object data type to date format. this i am doing it for order_date
df['order_date'] = pd.to_datetime(df['order_date'], format ="%Y-%m-%d")
```

```python
#drop columns - list price - cost price - discount percentage
df.drop(columns = ['list_price', 'cost_price', 'discount_percent'], inplace = True)
```
